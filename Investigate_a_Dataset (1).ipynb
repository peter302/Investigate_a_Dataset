```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.


# filling NAN values for non-umeric columns
df1=df.loc[:,df.dtypes=="object"]


for i in df1.columns:
    df1[i].fillna(df1[i].mode()[0], inplace=True)
    
    print(df1.isna().sum())
    
    
     
```

```python
# filling NAN values for numeric columns
df2=df.loc[:,df.dtypes !="object"].drop(['id'],axis=1)
df2
for i in df2.columns:
    df2[i].fillna(df2[i].mean(), inplace=True)
    
    print(df2.isna().sum())
    
```

```python
#Merge df1 and df2
df3 =  pd.concat([df1, df2], axis=1)
df3['id']= df['id']
df3
```

```python
##checking for duplicates

df3.duplicated()==True
```
